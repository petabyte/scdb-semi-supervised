```python
import pandas as pd
import transformers
```

```python
semi_training_df = pd.read_pickle('semi_super_training_data_with_use_vectors_80.zip', compression='zip')
semi_training_df.head(5)
```

```python
labeled_test_df = pd.read_pickle('training_data_with_use_vectors.zip', compression='zip')
labeled_test_df.head(5)
```

```python
from transformers import pipeline
summarizer = pipeline("summarization" , device=0)
summarizer("Sam Shleifer writes the best docstring examples in the whole world.", min_length=5, max_length=20)
```

```python
def summarize_case(text):
    return summarizer(text, min_length=150, max_length=500)
```

```python
import tqdm
from tqdm import tqdm
tqdm.pandas()
semi_training_df['summary'] = semi_training_df['text'].progress_apply(lambda x: summarize_case(x))
labeled_test_df['summary'] = labeled_test_df['text'].progress_apply(lambda x: summarize_case(x))
```

```python
semi_training_df.to_pickle('semi_super_training_data_with_use_vectors_80_with_summary.zip', compression='zip')
labeled_test_df.to_pickle('training_data_with_use_vectors_with_summary.zip', compression='zip')
```
